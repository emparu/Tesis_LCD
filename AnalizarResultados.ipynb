```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Supongamos que ya tienes los puntajes promedio calculados
Gemma_scores = [
    avg_score_Gemma_2B_hard_1, avg_score_Gemma_2B_hard_2, avg_score_Gemma_2B_hard_3,
    avg_score_Gemma_9B_hard_1, avg_score_Gemma_9B_hard_2, avg_score_Gemma_9B_hard_3,
    avg_score_Gemma_27B_hard_1, avg_score_Gemma_27B_hard_2, avg_score_Gemma_27B_hard_3
]

Gemini_scores = [
    avg_score_Gemini_2B_hard_1, avg_score_Gemini_2B_hard_2, avg_score_Gemini_2B_hard_3,
    avg_score_Gemini_9B_hard_1, avg_score_Gemini_9B_hard_2, avg_score_Gemini_9B_hard_3,
    avg_score_Gemini_27B_hard_1, avg_score_Gemini_27B_hard_2, avg_score_Gemini_27B_hard_3
]

# Cálculo de intervalos de confianza del 90%
# Cálculo de intervalos de confianza del 90%
n = 177  # Número de ejemplos
confidence_level = 0.90

# Cálculo del cuantil para la normal
z_score = stats.norm.ppf((1 + confidence_level) / 2)

# Cálculo de la desviación estándar para una distribución Bernoulli
# std = sqrt(p * (1 - p))
std_dev_Gemma = [np.sqrt(prob * (1 - prob)) for prob in Gemma_scores]
std_dev_Gemini = [np.sqrt(prob * (1 - prob)) for prob in Gemini_scores]



# Cálculo de los márgenes de error
margin_of_error_Gemma = [z_score * (std / np.sqrt(n)) for std in std_dev_Gemma]
margin_of_error_Gemini = [z_score * (std / np.sqrt(n)) for std in std_dev_Gemini]

# Etiquetas para el eje x
labels = [
    '2B_1_turno', '2B_2_turnos', '2B_3_turnos',
    '9B_1_turno', '9B_2_turnos', '9B_3_turnos',
    '27B_1_turno', '27B_2_turnos', '27B_3_turnos'
]

# Configurar el ancho de las barras y las posiciones de las barras
x = np.arange(len(labels))  # Las ubicaciones de las etiquetas
width = 0.35  # El ancho de las barras

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar las barras para Gemma y Gemini
rects1 = ax.bar(x - width/2, Gemma_scores, width, label='Gemma', color='skyblue')
rects2 = ax.bar(x + width/2, Gemini_scores, width, label='Gemini', color='lightcoral')

# Añadir intervalos de confianza
for i in range(len(Gemma_scores)):
    ax.errorbar(x[i] - width/2, Gemma_scores[i], yerr=margin_of_error_Gemma[i], 
                fmt='o', color='black', capsize=5)
    ax.errorbar(x[i] + width/2, Gemini_scores[i], yerr=margin_of_error_Gemini[i], 
                fmt='o', color='black', capsize=5)

# Añadir etiquetas, título y etiquetas personalizadas para el eje x
ax.set_xlabel('Modelo y Turnos')
ax.set_ylabel('Accuracy')
#ax.set_title('Accuracy de los jueces 2B, 9B y 27B para debates generados con Gemma y Gemini con intervalos de confianza del 90%')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.legend()

# Guardar el gráfico como un archivo JPG
plt.tight_layout()
plt.savefig('puntuaciones_promedio_gemma_vs_gemini.png', dpi=300)
plt.show()
```
